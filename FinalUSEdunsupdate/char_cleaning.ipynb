```python
import pandas as pd 
import openpyxl
import xlrd
```

```python
df = pd.read_excel(r"C:\Users\ck261\Documents\IES\Source Files\ProjDesc.xlsx")
```

```python
def clean(col):
    df[col] = df[col].str.replace(";#","")
    df[col] = df[col].str.replace("XX_","")
    df[col] = df[col].str.replace("\d+", ",")
    df[col] = df[col].str.replace(" and ",",")
    df[col] = df[col].str.replace("_","")
    df[col] = df[col].str.rstrip(",")
    df[col]
```

```python
clean("Data source")
clean("Goal")
clean("Data collection")
clean("Disability area")
clean("Project target")
clean("State(s)/Territories")
clean("Special population: Primary focus")
clean("Special population: Sub analysis")
clean("Student age")
clean("Student grade level")

df
```
